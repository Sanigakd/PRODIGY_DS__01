#**Data Visualization**

```python
plt.figure(figsize=(9,9))
sns.heatmap(data.corr(),cmap='Blues',annot=True)
```

```python
# make a plot for Population Distribution by Continent
grouped_data = data.groupby('Continent').sum(numeric_only=True).sort_values(by="2022 Population", ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data['2022 Population'], color='#66B2FF', edgecolor='black')
plt.xlabel('Continent')
plt.ylabel('Population (2022)')
plt.title('Bar Population Distribution by Continent 2022')
plt.xticks(rotation=60)
plt.grid(True)

# Adding data labels above each bar
for i, v in enumerate(grouped_data['2022 Population']):
    plt.text(i, v + 500000, str(v), ha='center', va='bottom', fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(6, 8))
labels = grouped_data.index.tolist()
colors=['#0000FF','#3399FF','#66B2FF','#CCE5FF','#E6F2FF','#99CCFF']
plt.pie(data['Continent'].value_counts(),autopct='%1.1f%%',colors=colors,labels=labels)
plt.legend(data['Continent'].value_counts().index,loc=(1.2,0.2))
```

```python
# Group the data by continent and calculate the sum of population for each year
grouped_data = data.groupby('Continent').agg({
    '2022 Population': 'sum',
    '2020 Population': 'sum',
    '2015 Population': 'sum',
    '2010 Population': 'sum',
    '2000 Population': 'sum',
    '1990 Population': 'sum',
    '1980 Population': 'sum',
    '1970 Population': 'sum'}).reset_index()

years = ['2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']

fig, axes = plt.subplots(2, 4, figsize=(10, 8), sharex=True, sharey=True)
axes = axes.ravel()

# Loop through the years and create a scatter plot for each continent
for i, year in enumerate(years):
    axes[i].scatter(grouped_data['Continent'], grouped_data[year + ' Population'])
    axes[i].set_title(year)
    axes[i].set_xlabel('Continent')
    axes[i].set_ylabel('Population')
    axes[i].set_xticks(range(len(grouped_data['Continent'])))
    axes[i].set_xticklabels(grouped_data['Continent'], rotation=90)

plt.tight_layout()
plt.show()
```

```python
# 'Population by Continent (1970-2022)'
df = {
    "Continent": data['Continent'],
    "1970 Population": data['1970 Population'],
    "1980 Population": data['1980 Population'],
    "1990 Population": data['1990 Population'],
    "2000 Population": data['2000 Population'],
    "2010 Population": data['2010 Population'],
    "2015 Population": data['2015 Population'],
    "2020 Population": data['2020 Population'],
    "2022 Population": data['2022 Population']}

population_by_continent = pd.DataFrame(df)

# Plotting
years = population_by_continent.columns[1:]
plt.figure(figsize=(12, 5))
for continent in population_by_continent["Continent"].unique():
    continent_data = population_by_continent[population_by_continent["Continent"] == continent].iloc[0, 1:]
    plt.plot(years, continent_data, label=continent)

plt.title('Population by Continent (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()

```

```python
# Plotting a horizontal bar chart to show the top 10 countries with the largest population in 2022
top_10_population_2022 = data.nlargest(10, '2022 Population')

plt.figure(figsize=(10, 6))
bars = plt.barh(top_10_population_2022['Country/Territory'], top_10_population_2022['2022 Population'], color='#CCE5FF', edgecolor='black')
plt.xlabel('Population (2022)')
plt.ylabel('Country/Territory')
plt.title('Top 10 Countries with Largest Population in 2022')
plt.grid(axis='x')

# Adding data labels on the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1000000, bar.get_y() + bar.get_height()/2, f'{width:,}', ha='left', va='center', fontsize=9)

plt.show()
```

```python
# Plot a line chart of the population growth rate for individual countries to identify trends.
x = data['Country/Territory'].head(20)
y = data['Growth Rate'].head(20)
plt.figure(figsize=(12, 5))
plt.plot(x, y, marker='h', markersize=8, color='#3399FF', linewidth=2)
plt.xlabel('Country/Territory')
plt.ylabel('Growth Rate')
plt.title('Population Growth Rate for 20 Countries')
plt.xticks(rotation=90)
plt.grid(color='#99CCFF')
```

```python
# Visualize the top 7 most densely populated countries on a world map

x = data['Country/Territory'].head(7)
y = data['Density (per km²)'].head(7)
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue', edgecolor='black')
plt.xlabel('Country/Territory')
plt.ylabel('Population Density (per km²)')
plt.title('Population Density Comparison by Country')
plt.xticks(rotation=90)
plt.grid(color='blue')
plt.show()

```

```python
top_countries = data.sort_values(by='2022 Population', ascending=False).head(10)
pivot_data = top_countries.pivot(index='Country/Territory', columns='Continent', values='2022 Population')

pivot_data.plot(kind='bar', stacked=True, colormap='tab20', edgecolor='black')
plt.xlabel('Country/Territory')
plt.ylabel('Population (2022)')
plt.title(f'Population of Top 10 Most Populated Countries by Continent (2022)')
plt.xticks(rotation=90)
plt.legend(title='Continent', loc='upper left', bbox_to_anchor=(1.1, 1))
plt.grid(color='blue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,7))

x = data['Country/Territory'].head(20)
y = data['2022 Population'].head(20)

# Plot the scatter plot with country names and numbers on y-axis
marker_sizes = data['2022 Population'] / 50000
for i, country in enumerate(x):
    plt.scatter(country, y.iloc[i], s=(marker_sizes.iloc[i])/20, label=country, alpha=0.7)

# Set y-axis to display numbers in billions
plt.ticklabel_format(style='plain', axis='y', useOffset=False, scilimits=(9, 9))

plt.xlabel('Country/Territory')
plt.ylabel('Population (2022)')
plt.title('Scatter Plot of 2022 Population for Most 20 populated countries')
plt.xticks(rotation=90)
plt.grid(color='lightblue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

x = data['Country/Territory'].head(20)
y = data['2022 Population'].head(20)

# Plot the scatter plot with country names and numbers on y-axis
marker_sizes = data['2022 Population'] / 50000
for i, country in enumerate(x):
    plt.scatter(country, y.iloc[i], s=(marker_sizes.iloc[i])/20, label=country, alpha=0.7)
    plt.text(country, y.iloc[i], f'{y.iloc[i]:,.0f}', ha='center', va='bottom', rotation='vertical', fontsize=10, color='black')

# Set y-axis to display numbers in billions
plt.ticklabel_format(style='plain', axis='y', useOffset=False, scilimits=(9, 9))

plt.xlabel('Country/Territory')
plt.ylabel('Population (2022)')
plt.title('Scatter Plot of 2022 Population for Most 20 populated countries')
plt.xticks(rotation=90)
plt.grid()
plt.tight_layout()
plt.show()
```
